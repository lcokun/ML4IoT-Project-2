```python
from pathlib import Path

import pandas as pd
from IPython.display import display

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

DATA_DIR = Path("data")

def read_head(path: Path, n: int = 5) -> pd.DataFrame:
    if path.suffix.lower() == ".csv":
        # mitbih_*/ptbdb_*.csv ship without a header row; header=None keeps
        # the first sample as data instead of misreading it as column names.
        return pd.read_csv(path, header=None, nrows=n)
    if path.suffix.lower() == ".xlsx":
        return pd.read_excel(path, nrows=n)
    raise ValueError(f"unsupported file type: {path.suffix}")

for path in sorted(DATA_DIR.rglob("*")):
    if path.suffix.lower() not in (".csv", ".xlsx"):
        continue

    print(f"=== {path.relative_to(DATA_DIR)} ===")
    display(read_head(path))
```
